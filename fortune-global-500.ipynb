```python
import pandas as pd
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Fortune_Global_500"

tables = pd.read_html(url)
for i, tbl in enumerate(tables):
    print(i, tbl.shape)
    print(tbl.columns.tolist())
    print()

df = tables[1]
```

```python
df = tables[24]
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r"\s+", "_", regex=True)
      .str.replace(r"[^\w_]", "", regex=True)
)

df = df.dropna(how="all", axis=0).dropna(how="all", axis=1)

df = df.reset_index(drop=True)

print(df.head())
print(df.info())
```

```python
output_csv = "fortune_global_500_2023.csv"
df.to_csv(output_csv, index=False)
print(f"Saved cleaned dataset to {output_csv}")
```
